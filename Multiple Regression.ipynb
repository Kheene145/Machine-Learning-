## Multiple Regression

```python
## Multiple regression is like linear regression, but with more than one independent value, meaning that we try to predict
## a value based on two or more variables.
```

```python
import pandas as pd

df = pd.read_csv("data.csv")
```

```python
df
```

```python
X = df[["Weight", "Volume"]]
y = df["CO2"]
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()

regr.fit(X, y)
```

```python
# predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1300cm^3

X = df[["Weight", "Volume"]]
y = df["CO2"]

regr = linear_model.LinearRegression()

regr.fit(X, y)

predictedCO2 = regr.predict([[2300,1300]])

print(predictedCO2)
```

```python
## Coefficient
```

```python
# It is a factor that describe the relationship with an unknown variable.

# e.g The coefficient of x in 2x is 2. 
```

```python
# The coefficient values of weight against CO2, and for volume against CO2. The answers we get tells us what would happen

# if we increase or decrease one of the independent value.
```

```python
# e.g print the coefficient values of the regression object:


print(regr.coef_)
```

```python
predictedCO2 = regr.predict([[3300, 1300]])

print(predictedCO2)
```
