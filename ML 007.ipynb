## Linear Regression

```python
# It uses the relationship between the data-points to draw a straight line through all of them. 
# This line can be used to predict future values.
```

```python
import matplotlib.pyplot as plt

x = [8,9,6,7,3,8,9,4,7,3,2,8,9]

y = [16,98,78,66,54,33,52,76,134,54,87,45,34]

plt.scatter(x,y)
plt.show()
```

```python
# import scipy and draw the line of Linear Regression.

import matplotlib.pyplot as plt
from scipy import stats
```

```python
x = [8,9,6,7,3,8,9,4,7,3,2,8,9]

y = [16,98,78,66,54,33,52,76,134,54,87,45,34]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plt(x, mymodel)
plt.show()
```
