### This dataset is highly unbalance

```python
# separating the data for analysis

legit = creditcard_df[creditcard_df.Class == 0]
fraud = creditcard_df[creditcard_df.Class == 1]
```

### Under_Sampling

#### Build a sample dataset containing similar distribution of normal transaction and fraudulent transaction

#### Number of Fraudulent Transactions = 492

```python
legit_sample = legit.sample(n=492)
```

```python
# Concatenate two dataframes

new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
### Splitting the data into Features and Target
```

```python
X = new_dataset.drop(columns= 'Class', axis=1)
y = new_dataset['Class']
```

```python
print(X)
```

```python
print(y)
```

```python
#### Splitting the data into Training and testing data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Training Model
# using Logistic Regression best approach for binary output variable
```

```python
from sklearn.preprocessing import StandardScaler

model = LogisticRegression(max_iter=500)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model.fit(X_train_scaled, y_train)
```

```python
model.coef_
```

```python
print(model)
```

### Model Evaluation
#### Accuracy Score 

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
```

```python
print(f"Model Accuracy: {accuracy:.4f}")
```
