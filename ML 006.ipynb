## Scatter Plot

```python
# A matplotlib module where each value in the data set is represented by a dot, it needs a two array of the same length, one for
# the x-axis and one for the y-axis.
```

```python
x = [8,9,6,7,3,8,9,4,7,3,2,8,9]

y = [16,98,78,66,54,33,52,76,134,54,87,45,34]

# The x array represents the age of each car
# the y array represents the speed of each car
```

```python
import matplotlib.pyplot as plt
```

```python
# Use the scatter() method to draw a scatter plot diagram:

x = [8,9,6,7,3,8,9,4,7,3,2,8,9]

y = [16,98,78,66,54,33,52,76,134,54,87,45,34]

plt.scatter(x, y)
plt.show()
```

```python
# A scatter plot with 1000 dots:

import numpy
import matplotlib.pyplot as plt

x = numpy.random.normal(5.0, 1.0, 1000)
y = numpy.random.normal(10.0, 2.0,1000)

plt.scatter(x,y)
plt.show()
```
